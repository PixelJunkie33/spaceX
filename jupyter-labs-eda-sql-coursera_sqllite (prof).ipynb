##### Generate a query to display the names of the unique launch sites  in the space mission


```python
launch_sites = df['Launch_Site'].unique()

# Print the unique launch sites
print('Unique launch sites:')
for site in launch_sites:
    print(site)
```



#### Create a query to find and display the total payload mass carried by boosters launched by NASA (CRS)


```python
#display the total number of boosters launched by NASA CRS
import pandas as pd
filtered_df = df.loc[df['Customer'] == 'NASA (CRS)']
Total_Payload_Mass_Carried= filtered_df['PAYLOAD_MASS__KG_'].sum()
print('Total Payload Mass Carried by NASA (CRS):', Total_Payload_Mass_Carried, 'kg')
```




##### Query the average payload mass carried by booster version F9 v1.1


```python
#display average payload mass carried by boosters version f9 v1.1
df1 = df.loc[df['Booster_Version'] == 'F9 v1.1']

#filter the data to only include average payload mass
df1 = df1['PAYLOAD_MASS__KG_'].mean()

print('Average Payload Mass Carried by F9 v1.1:', df1, 'kg')
```



##### Here we find the date when the first successful landing outcome in ground pad was achieved.



```python
import pandas as pd

df2 = df.loc[df["Landing _Outcome"] == 'Success (ground pad)']
df2 = df2['Date']

print('First Landing Success (ground pad):', df2.iloc[0])
```

##### Create a query to generate a list with the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python

df3 = df.loc[(df["Landing _Outcome"] == 'Success (drone ship)') & (df["PAYLOAD_MASS__KG_"] > 4000) & (df["PAYLOAD_MASS__KG_"] < 6000)]

df3 = df3[['Booster_Version', 'PAYLOAD_MASS__KG_', 'Landing _Outcome']]
print('Successful Landing Outcomes for Payload Mass between 4000 and 6000 kg:')
df3 = pd.DataFrame(df3)
df3
```


##### Query a list count of the total number of successful and failure mission outcomes


```python
#list the total number of sucessful and failure mission outcomes
df4 = df['Mission_Outcome'].value_counts()
df4 = pd.DataFrame(df4)
df4


```


#### Subquery a list with the names of the booster_versions which have carried the maximum payload mass.


```python
df5 = df.loc[df['PAYLOAD_MASS__KG_'] == df['PAYLOAD_MASS__KG_'].max()]
df5 = df5[['Booster_Version', 'PAYLOAD_MASS__KG_']]
df5= pd.DataFrame(df5)
df5
```


#### Create a query to generate a list of records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.


```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month_name()

df6 = df.loc[(df['Year'] == 2015) & (df['Landing _Outcome'] == 'Failure (drone ship)')]
df6 = df6[['Month','Year', 'Landing _Outcome', 'Booster_Version', 'Launch_Site']]
df6
```
